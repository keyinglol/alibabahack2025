# Holidays

```python
import holidays

# Get Malaysian holidays for the years in your dataset
# First, determine the range of years in your data
min_year = int(df['year'].min())
max_year = int(df['year'].max())

# Create a Malaysian holiday calendar
my_holidays = holidays.US(years=range(min_year, max_year + 1))

# Convert transaction_date to datetime if it's not already
if not pd.api.types.is_datetime64_any_dtype(df['transaction_date']):
    df['transaction_date'] = pd.to_datetime(df['transaction_date'])

# Check if each date is a holiday
df['is_public_holiday'] = df['transaction_date'].apply(lambda x: 1 if x in my_holidays else 0)

# If you want the holiday name as well
df['holiday_name'] = df['transaction_date'].apply(lambda x: my_holidays.get(x) if x in my_holidays else None)
```

```python
# Step 1: Merge df with product_info to get unit prices
merged = df.merge(product_info[['product_id', 'unit_price']], on='product_id', how='left')

# Step 2: Calculate RM value per row (transaction line)
merged['transaction_value'] = merged['transaction_qty'] * merged['unit_price']

# Step 3: Group by date and sum
daily_total_sales = (
    merged
    .groupby('transaction_date')['transaction_value']
    .sum()
    .reset_index()
    .rename(columns={'transaction_value': 'daily_total_sales_usd'})
)

# daily total sales truncate to 2 decimal
daily_total_sales['daily_total_sales_usd'] = daily_total_sales['daily_total_sales_usd'].apply(lambda x: "{:.2f}".format(x))

daily_total_sales
```

# Simulate inventory

```python
import pandas as pd
import numpy as np

# Assume daily_sales is already created and sorted
daily_sales = daily_sales.sort_values(['product_id', 'transaction_date'])

# Create an empty list to store rows
inventory_records = []

# Simulate per product
for product_id, group in daily_sales.groupby('product_id'):
    previous_closing_stock = 100  # start with a fixed opening stock
    max_wastage = 30  # initial high wastage to simulate improvement
    for idx, row in group.iterrows():
        date = row['transaction_date']
        units_sold = row['transaction_qty']
        
        # Simulate purchase quantity (could be random or fixed, here we use fixed + small noise)
        purchases = np.random.randint(10, 30)
        
        # Simulate wastage reducing over time
        wastage = max(0, int(np.random.uniform(0, max_wastage)))
        max_wastage = max(0, max_wastage + 0.1)  # slowly decrease allowed max wastage
        
        # Calculate available stock
        available_stock = previous_closing_stock + purchases
        
        # Calculate closing stock
        closing_stock = available_stock - units_sold - wastage
        
        # Ensure closing stock is not negative
        if closing_stock < 0:
            closing_stock = 0
            wastage = available_stock - units_sold  # adjust wastage accordingly
            wastage = max(0, wastage)  # make sure wastage is not negative again
        
        inventory_records.append({
            'transaction_date': date,
            'product_id': product_id,
            'opening_stock': previous_closing_stock,
            'purchases': purchases,
            'closing_stock': closing_stock,
            'units_sold': units_sold,
            'wastage': wastage
        })
        
        # Next day's opening stock is today's closing
        previous_closing_stock = closing_stock

# Create DataFrame
inventory_daily = pd.DataFrame(inventory_records)

# sort by transaction date, then nested by sort by product id
inventory_daily = inventory_daily.sort_values(['transaction_date', 'product_id']).reset_index(drop=True)
inventory_daily
```

```python
import matplotlib.pyplot as plt

avg_wastage = inventory_daily.groupby('transaction_date')['wastage'].mean()

plt.plot(avg_wastage.index, avg_wastage.values)
plt.title("Average Daily Wastage (Simulated)")
plt.xlabel("Date")
plt.ylabel("Average Wastage Units")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
